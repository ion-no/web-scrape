```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://editorial.rottentomatoes.com/guide/best-horror-movies-of-all-time/"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
divs = soup.find_all('div', {'class' : 'col-sm-18 col-full-xs countdown-item-content'})
```

```python
divs[0].find('h2')
```

```python
headings = [div.find('h2') for div in divs]
```

```python
#Title

[heading.find('a') for heading in headings]
```

```python
movie_names = [heading.find('a').string for heading in headings]
movie_names
```

```python
#Year

[heading.find('span', {'class' : 'subtle start-year'}) for heading in headings]
```

```python
movie_years = [heading.find('span', {'class' : 'subtle start-year'}).string[1 : -1] for heading in headings]
```

```python
movie_years = [int(movie_year) for movie_year in movie_years]
movie_years
```

```python
headings[0]
```

```python
#Ratings

movie_rating = [heading.find('span', {'class' : 'tMeterScore'}).string for heading in headings]
```

```python
movie_rating = [rate.strip('%') for rate in movie_rating]
movie_rating
```

```python
movie_rating = [int(rate) for rate in movie_rating]
```

```python
director = [div.find('div', {'class' : 'info director'}) for div in divs]
director[0]
```

```python
directors = [dir.find('a').string for dir in director]
directors
```

```python
#Casts

casting = [div.find('div', {'class' : 'info cast'}) for div in divs]
casting
```

```python
casts = [castings.find_all('a') for castings in casting]
casts
```

```python
cast = [c.string for c in casts[0]]
cast = ', '.join(cast)
cast
```

```python
all_casts = []

for c in casts:
    cast_link = [x.string for x in c]
    cast_name = [link.string for link in cast_link]
    all_casts.append(', '.join(cast_name))
```

```python
all_casts
```
